```python
from requests import Request, Session
from requests.exceptions import ConnectionError, Timeout, TooManyRedirects
import json
import os

# Ladda API-nyckeln från miljövariabel
from dotenv import load_dotenv

load_dotenv()
API_KEY = os.getenv("COINMARKETCAP_API")

# API URL och parametrar
url = "https://pro-api.coinmarketcap.com/v1/cryptocurrency/quotes/latest"
parameters = {
    'symbol': 'BTC',  # Hämtar data för Bitcoin
    'convert': 'USD'
}
headers = {
    'Accepts': 'application/json',
    'X-CMC_PRO_API_KEY': API_KEY,
}

# Skapa en session och ställ in headers
session = Session()
session.headers.update(headers)

# Hämta data från API:et
try:
    response = session.get(url, params=parameters)
    data = json.loads(response.text)
    print(json.dumps(data, indent=4))  # Skriv ut data i ett läsbart format
except (ConnectionError, Timeout, TooManyRedirects) as e:
    print("Fel vid API-anrop:", e)
```
